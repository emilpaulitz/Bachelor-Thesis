```python
N = 10
```

```python
%%time
from pycolator import *

idCol = 'SpecId'
scoreCol = 'NuXL:score'
excluded = ['SpecId', 'Label', 'ScanNr', 'Peptide', 'Proteins', 'FDR', 'q-val', 'class-specific_q-val', 'Rank', 'temp_score', 'percolator_score']

dSlow = readAndProcess('../data/1-AChernev_080219_dir_HeLa_cyt_UCGA_slow.tsv', idCol, scoreCol, excludedCols = excluded)
features = [x for x in list(dSlow.columns) if (x not in excluded)]
```

```python
%%time

saveName = ''

optimalRanking = []

for i in range(N):
    p = percolator_experimental(dSlow, idCol, features, plotSaveName = saveName)
    optimalRanking.append([pseudoROC(p, plot = False), pseudoROC(p.loc[p['NuXL:isXL'] == 1], plot = False), pseudoROC(p.loc[p['NuXL:isXL'] == 0], plot = False)])
```

```python
%%time

saveName = ''

noRanking = []

for i in range(N):
    p = percolator_experimental(dSlow, idCol, features, optimalRanking = False, plotSaveName = saveName)
    noRanking.append([pseudoROC(p, plot = False), pseudoROC(p.loc[p['NuXL:isXL'] == 1], plot = False), pseudoROC(p.loc[p['NuXL:isXL'] == 0], plot = False)])
```

```python
%%time

saveName = ''

rankOne = []

for i in range(N):
    p = percolator_experimental(dSlow, idCol, features, optimalRanking = False, useRankOneOnly = True ,plotSaveName = saveName)
    rankOne.append([pseudoROC(p, plot = False), pseudoROC(p.loc[p['NuXL:isXL'] == 1], plot = False), pseudoROC(p.loc[p['NuXL:isXL'] == 0], plot = False)])
```

```python
import csv

with open('../results/optimalRanking_new.csv', mode='w') as csvFile:
    csvWriter = csv.writer(csvFile, delimiter='\t', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    
    csvWriter.writerow(['optimal Ranking'])
    for r in optimalRanking:
        csvWriter.writerow([str(x) for x in r])
    csvWriter.writerow(['all Ranks'])
    for r in noRanking:
        csvWriter.writerow([str(x) for x in r])
    csvWriter.writerow(['Rank one'])
    for r in rankOne:
        csvWriter.writerow([str(x) for x in r])
```
