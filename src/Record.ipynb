```python
%%time
from pycolator import *

idCol = 'SpecId'
scoreCol = 'NuXL:score'
excluded = ['SpecId', 'Label', 'ScanNr', 'Peptide', 'Proteins', 'FDR', 'q-val', 'class-specific_q-val', 'Rank', 'temp_score', 'percolator_score']

dSlow = readAndProcess('../data/1-AChernev_080219_dir_HeLa_cyt_UCGA_slow.tsv', idCol, scoreCol, excludedCols = excluded)
features = [x for x in list(dSlow.columns) if (x not in excluded)]
```

```python
%%time 

# Experiment: mit den verschiedenen Klassen je einzeln trainieren
c = dSlow.loc[dSlow['NuXL:XL_C'] == 1]
a = dSlow.loc[dSlow['NuXL:XL_A'] == 1]
g = dSlow.loc[dSlow['NuXL:XL_G'] == 1]
u = dSlow.loc[dSlow['NuXL:XL_U'] == 1]
none = dSlow.loc[dSlow['NuXL:isXL'] == 0]

saveName = ''

pc = percolator_experimental(c, idCol, features, plotSaveName = saveName, plotXLnXL = False)
pa = percolator_experimental(a, idCol, features, plotSaveName = saveName, plotXLnXL = False)
pg = percolator_experimental(g, idCol, features, plotSaveName = saveName, plotXLnXL = False)
pu = percolator_experimental(u, idCol, features, plotSaveName = saveName, plotXLnXL = False)
pn = percolator_experimental(none, idCol, features, plotSaveName = saveName, plotXLnXL = False)
```

```python
import pycohelper as ph
pseudoROC(pc, label = 'c')
pseudoROC(pa, label = 'a')
pseudoROC(pg, label = 'g')
pseudoROC(pu, label = 'u')
pseudoROC(pn, label = 'none')
p = pd.concat([pc,pa,pg,pu,pn])
p = ph.calcQ(p, 'q-val', addXlQ = False, ascending = True)
pseudoROC(p, label = 'total')
plt.savefig('Test.png')
```
